```python
def Movie_rank(file):
     
    f=open(file,'r',encoding='utf8')
    fp = f.read()
    f.close()
    
    m=fp.split('\n')
    
    
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np

    from matplotlib import font_manager, rc


    #한글을 읽을 수 있도록
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/MALGUN.TTF").get_name()
    rc('font', family=font_name)
    
    s1=np.array([float(x.split(',  ')[1]) for x in m])  #숫자
    s2=np.array([float(x.split(',  ')[2]) for x in m] ) #숫자
    s3=np.array([float(x.split(',  ')[-1]) for x in m])  #숫자
    
    n=[x.split(',  ')[0] for x in m]  #글자
    
    x=np.arange(0,18,1.5)
    
    plt.figure(figsize=(100,50))
    plt.bar(x,s1, color='r', width=0.3, label='관람객')  #width는 bar두께 label은 bar이름
    plt.bar(x+0.3,s2, color='g', width=0.3,label='기자&평론가')
    plt.bar(x+0.6,s3, color='b', width=0.3,label='네티즌')
    
    plt.ylabel('score')
    plt.xlabel('who')
    
    plt.legend(loc=2)
    
    plt.xticks(x+0.3,n)
    
    plt.show()
    
     
```

```python
file='./movie_rank.txt'
Movie_rank(file)
```
