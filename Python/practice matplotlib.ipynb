## pie그래프  원그래프

## fig1, ax1 = plt.subplots()

## color=('b','g','r','c','m','y','k','w')    
 colors : A sequence of matplotlib color args through which the pie chart

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes  = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  #떨어진 만큼
# only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## 비율 나타내기
## int(숫자/100.*np.sum(명칭))

```python
recipe = ["375 g 밀가루",
          "75 g  설탕",
          "250 g 버터",
          "300 g 딸기"]
```

```python
fig, ax=plt.subplots(figsize=(10,12),subplot_kw=dict(aspect="equal"))  #원그래프 크기


data=[float(x.split()[0]) for x in recipe]  #숫자
ingredients=[x.split()[-1] for x in recipe]  #글자


def func(a,b):
    absolute=int(a/100.*np.sum(b))
    return '{:.1f}%\n({:d} g)'.format(a,absolute)


wedges,texts,autotexts=ax.pie(data, autopct=lambda a: func(a, data),
                                  textprops=dict(color="w"))
#autopct는 func()의 값을 가져온다.
#textprops=dict(color="글자색깔")


ax.legend(wedges, ingredients,
          title="주요성분",
          loc=1,  # EX) loc="center left"
          bbox_to_anchor=(1, 0, 0.5, 1))


plt.setp(autotexts, size=30, weight="bold")  #글자 크기

ax.set_title("딸기케잌")

plt.show()
```

## 다중 곡선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.linspace(0,2*np.pi,100)
Ya=np.cos(X)
Yb=1/2*np.tan(X)

plt.plot(X,Ya) #blue
plt.plot(X,Yb)  #orange
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X,Y):   #기울기
    Xs=X[1:]-X[:-1]
    Ys=Y[1:]-Y[:-1]
    plt.plot(X[1:],Ys/Xs)
    
plt.plot(X,Yb)
plot_slope(X,Yb)
plt.show()
```

## 다중 막대 차트 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(5, 2)


color_list = ['b', 'g', 'r','y','k']

gap = .8 / len(data)

for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i*gap, row, 
           width = gap, 
           color = color_list[i % len(color_list)])
    
plt.grid()

plt.show()
```
