```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

```python
boston=load_boston()
print(boston.keys())
print(boston.items())
```

```python
x=boston.data
y=boston.target
len(x), len(y)
```

## 데이터셋 나누기

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,   #학습 75%, value data 25%
                                                  test_size=0.25,
                                               random_state=42)
```

## 모델 만들어 학습 및 예측

#### 1. 모델 만들기
#### 2. 모델 학습  model.fit()
#### 3. 모델 예측 model.predict()
#### 4. 모델 평가

#### 범주형일 때는 knn을 사용
#### 수치형일 때는 LinearRegression을 사용

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(x_train,y_train)
pred=model.predict(x_test)
pred
```

```python
import pandas as pd
df=pd.DataFrame({"y_test" : y_test,"prediction" : pred, "diff" : (y_test - pred)})
df
```

```python
#MAE, MSE 가 낮을 수록 좋다.

MAE=sum(abs(df["diff"]))/len(df["diff"])
MSE=sum((df["diff"])**2)/len(df["diff"])

import numpy as np

RMSE=np.sqrt(MSE)

print(MAE)
print(MSE)
print(RMSE)
```

```python
acc_tr=model.score(x_train,y_train)
acc_test=model.score(x_test,y_test)
print("훈련 세트의 정확도 : {:.2f}".format(acc_tr))  
print("테스트 세트의 정확도 : {:.2f}".format(acc_test))  
```
